```python
df.describe()
```

```python
df.head()
```

```python
### columns to clean
cols=['Installs','Price']

### characters to remove
chars=['+',',','$']

# storing columns_to_clean in col variable
for col in cols:
    #storing characters_to_clean in char varaible
    for char in chars:
        # Replacing unnecessary characters with empty string
        df[col]=df[col].apply(lambda x: x.replace(char,''))

# check a random sample of head of 5 rows
df.head()
```

### Procedure for converting the column "Size" to float
#### There are sizes counted in mb, kb, in numbers without measurement unit and with "varies with device"

```python
#removing the 'M'
df.Size=df.Size.str.replace('M','')
```

```python
#removing the 'Varies with device'
df.Size=df.Size.str.replace('Varies with device','')
```

Here we replace K and change the unit to Mb

```python
df.Size=df.Size.apply(lambda x: str(round((float(x.rstrip('K'))/1024),1))
                     if x[-1]=='K'
                     else x)
```

```python
df.Size.unique()
```

As the columns of **Genres** is same as the column for **category**, we drop the column for **Genres**.

Also, the column of **android version, current version, Last updated** are not of our use, so we drop these columns.

```python
df.drop(['Genres','Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```
